```python
import pandas as pd
```

```python
datos = pd.read_csv('animales.csv')
datos
```

```python
features = [[7, 0.6, 40],[7, 0.6, 41],[37, 0.8, 37],[37, 0.8, 38]]
```

```python
etiquetas = [0,0,1,1]
```

```python
from sklearn import tree
```

```python
clasificador = tree.DecisionTreeClassifier()
```

```python
clasificador.fit(features,etiquetas)
```

```python
predicion = clasificador.predict([[20, 0.7, 42]])
if (predicion == 0):
    print("los datos ingresados corresponden a un gato")
else:
  print("los datos ingresados corresponden a un perro")
```

```python
altura = int(input("ingrese la altura "))
peso = float(input("ingrese el peso "))
temperatura = float(input("ingrese la tempeatura "))

predicion = clasificador.predict([[altura, peso, temperatura]])
if (predicion == 0):
    print("los datos ingresados corresponden a un sos")
else:
  print("los datos ingresados corresponden a un sas")
```
